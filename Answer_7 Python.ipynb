Draw plots to demonstrate the analysis for the following questions for better visualizations
Write code comments wherever required for code understanding
Insights to be drawn - ● Get all the Earth meteorites that fell before the year 2000 ● Get all the earth meteorites co-ordinates who fell before the year 1970 ● Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg

```python
#Q7
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the meteorite data into a pandas DataFrame
df = pd.read_csv('meteorite_data.csv')

# Convert the year column to datetime type for easier filtering
df['year'] = pd.to_datetime(df['year'], errors='coerce')
```

```python
# Get all the Earth meteorites that fell before the year 2000
earth_meteorites_before_2000 = df[(df['reclat'].notnull()) & (df['year'] < '2000-01-01')]
earth_meteorites_before_2000
```

```python
# Get all the Earth meteorites' coordinates that fell before the year 1970
earth_meteorites_before_1970 = df[(df['reclat'].notnull()) & (df['year'] < '1970-01-01')]
earth_meteorites_before_1970
```

```python
# Get all Earth meteorites with a mass greater than 10000 kg (assuming mass is in kg)
mass_threshold = 10000
earth_meteorites_mass_greater_than_10000 = df[(df['reclat'].notnull()) & (df['mass'] > mass_threshold)]
earth_meteorites_mass_greater_than_10000
```

```python
# Plotting the results
sns.set(style="darkgrid")

# Plot 1: Earth meteorites that fell before 2000
plt.figure(figsize=(10, 6))
sns.scatterplot(data=earth_meteorites_before_2000, x='reclong', y='reclat', color='blue')
plt.title('Earth Meteorites that Fell Before 2000')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Plot 2: Earth meteorites' coordinates that fell before 1970
plt.figure(figsize=(10, 6))
sns.scatterplot(data=earth_meteorites_before_1970, x='reclong', y='reclat', color='green')
plt.title('Earth Meteorites\' Coordinates that Fell Before 1970')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Plot 3: Earth meteorites with mass greater than 10000 kg
plt.figure(figsize=(10, 6))
sns.scatterplot(data=earth_meteorites_mass_greater_than_10000, x='reclong', y='reclat', color='red')
plt.title('Earth Meteorites with Mass > 10000 kg')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
